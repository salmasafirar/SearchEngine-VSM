```python
def token_stopwords_stem(x):
  # Menghilangkan Stopwords
  stop_factory = StopWordRemoverFactory().get_stop_words()
  dictionary = ArrayDictionary(stop_factory)

  str1 = StopWordRemover(dictionary)
  str2 = str1.remove(x)
  
  # Stemming dengan library Sastrawi
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()
  katadasar = stemmer.stem(str2)

  # Tokenisasi
  hasil = katadasar.translate(str.maketrans('','',string.punctuation)).lower()
  hasil = word_tokenize(hasil)
  
  return hasil

ed_query = token_stopwords_stem(query)
list_kalimat1 = token_stopwords_stem(dc.iloc[0, 1])
list_kalimat2 = token_stopwords_stem(dc.iloc[1, 1])
list_kalimat3 = token_stopwords_stem(dc.iloc[2, 1])
list_kalimat4 = token_stopwords_stem(dc.iloc[3, 1])
list_kalimat5 = token_stopwords_stem(dc.iloc[4, 1])
list_kalimat6 = token_stopwords_stem(dc.iloc[5, 1])
list_kalimat7 = token_stopwords_stem(dc.iloc[6, 1])
list_kalimat8 = token_stopwords_stem(dc.iloc[7, 1])
list_kalimat9 = token_stopwords_stem(dc.iloc[8, 1])
list_kalimat10 = token_stopwords_stem(dc.iloc[9, 1])
list_kalimat11 = token_stopwords_stem(dc.iloc[10, 1])
list_kalimat12 = token_stopwords_stem(dc.iloc[11, 1])
list_kalimat13 = token_stopwords_stem(dc.iloc[12, 1])
list_kalimat14 = token_stopwords_stem(dc.iloc[13, 1])
list_kalimat15 = token_stopwords_stem(dc.iloc[14, 1])
list_kalimat16 = token_stopwords_stem(dc.iloc[15, 1])
list_kalimat17 = token_stopwords_stem(dc.iloc[16, 1])
list_kalimat18 = token_stopwords_stem(dc.iloc[17, 1])
list_kalimat19 = token_stopwords_stem(dc.iloc[18, 1])
list_kalimat20 = token_stopwords_stem(dc.iloc[19, 1])

print(ed_query)
print(list_kalimat1)
print(list_kalimat2)
print(list_kalimat3)
print(list_kalimat4)
print(list_kalimat5)
print(list_kalimat6)
print(list_kalimat7)
print(list_kalimat8)
print(list_kalimat9)
print(list_kalimat10)
print(list_kalimat11)
print(list_kalimat12)
print(list_kalimat13)
print(list_kalimat14)
print(list_kalimat15)
print(list_kalimat16)
print(list_kalimat17)
print(list_kalimat18)
print(list_kalimat19)
print(list_kalimat20)
```

##Frequency of Words
$tf = 1 + \log(f_{i, j})$

```python
def freq_words(list_word, cmb): #cmb = combined words
  freq = []
  for i in range(len(cmb)):
    if cmb[i] in list_word:
      counter = 0
      for j in range(len(list_word)):
        # count f i,j
        if cmb[i] == list_word[j]:
          counter+=1
      freq.append(1 + math.log(counter,10))
    else:
      freq.append(0)
  return freq

print(freq_words(ed_query, combine_words(docs, len(docs))))
print(freq_words(list_kalimat1, combine_words(docs, len(docs))))
print(freq_words(list_kalimat2, combine_words(docs, len(docs))))
print(freq_words(list_kalimat3, combine_words(docs, len(docs))))
print(freq_words(list_kalimat4, combine_words(docs, len(docs))))
print(freq_words(list_kalimat5, combine_words(docs, len(docs))))
print(freq_words(list_kalimat6, combine_words(docs, len(docs))))
print(freq_words(list_kalimat7, combine_words(docs, len(docs))))
print(freq_words(list_kalimat8, combine_words(docs, len(docs))))
print(freq_words(list_kalimat9, combine_words(docs, len(docs))))
print(freq_words(list_kalimat10, combine_words(docs, len(docs))))
print(freq_words(list_kalimat11, combine_words(docs, len(docs))))
print(freq_words(list_kalimat12, combine_words(docs, len(docs))))
print(freq_words(list_kalimat13, combine_words(docs, len(docs))))
print(freq_words(list_kalimat14, combine_words(docs, len(docs))))
print(freq_words(list_kalimat15, combine_words(docs, len(docs))))
print(freq_words(list_kalimat16, combine_words(docs, len(docs))))
print(freq_words(list_kalimat17, combine_words(docs, len(docs))))
print(freq_words(list_kalimat18, combine_words(docs, len(docs))))
print(freq_words(list_kalimat19, combine_words(docs, len(docs))))
print(freq_words(list_kalimat20, combine_words(docs, len(docs))))
```

##Count Idf
$idf = \log(\frac{N-n_i}{n_i})$

```python
def idf(dfx, n): # n = banyaknya dokumen
  dfx["dfi"] = dfx.iloc[:, 2:n+2].sum(axis=1)
  dfx["D/dfi"] = len(docs)/dfx["dfi"]
  dfx["Idf"] = dfx.apply(lambda x: math.log(x["D/dfi"]-1,10), axis=1)
  return dfx

idf(df, len(docs)).head()
```

##Weight

```python
def weight(dfx, n):
  wei = []
  wei1 = []
  idf = dfx['Idf']
  for i in range(1, n+2):
    wei = list(dfx.iloc[:, i] * idf)
    wei1.append(wei)

  for i in range(n+1):
    if i == 0:
      df['weight_query'] = wei1[i]
    else:
      df['weight_d'+str(i)] = wei1[i]
  return dfx

weight(df, len(docs)).head()
```

##Vector Length


```python
def panjang_vector(df,n):
  # Untuk mengetahui semua kolom pada data frame df 
  cols = df.columns
  # Untuk mengetahui kolom weight query dan dokumen2 inputan dimasukkan ke list_col_weight
  list_col_weight=[]
  for i in range(0,len(cols)):
    if(i>4+n):
      list_col_weight.append(cols[i])

  # Untuk menghitung panjang vector query dan dokumen yang hasilnya dimasukkan pada 
  weight = []
  result = []
  vector = 0
  for i in range(0,len(list_col_weight)):
    weight = df.loc[:, list_col_weight[i]]
    for i in weight:
      vector += i**2
    vector = math.sqrt(vector)
    result.append(vector)
    vector = 0

  return result

print(panjang_vector(df,len(docs)))
```

##Dot Product

```python
def dotproduct(df,n):
  # Untuk mengetahui semua kolom pada data frame df 
  cols = df.columns
  # Untuk mengetahui kolom weight query dan dokumen2 inputan dimasukkan ke list_col_weight
  list_col_weight=[]
  for i in range(0,len(cols)):
    if(i>4+n):
      list_col_weight.append(cols[i])

  # Untuk menghitung dot product query dan dokumen yang hasilnya dimasukkan pada 
  weight = []
  dot_product = []
  hasil = 0
  z = len(list_col_weight)-1
  for i in range(0,z):
    q = df.loc[:, list_col_weight[0]]
    d = df.loc[:, list_col_weight[i+1]]
    for i in range(0,len(q)):
      hasil += q[i]*d[i]
    dot_product.append(hasil)
    hasil = 0
  return dot_product

print(dotproduct(df,len(docs)))
```

##Similarity

```python
vector_length = panjang_vector(df,len(docs))
dot_product = dotproduct(df,len(docs))
query_length = vector_length[0]
def similarity(vector, dot, n): # n = len(docs)
  query_length = vector[0]
  rank = []
  for i in range(n):
    rank.append(dot[i]/(query_length*vector[i+1]))
  return rank

similarity(vector_length, dot_product, len(docs))
```
